```python
import pandas as pd
import numpy as np
from sklearn.cross_validation import train_test_split
train_data = pd.read_csv("train.csv")
train_data.shape
```

```python
train_data.head()
```

```python
print train_data.count()

train_data=train_data.dropna()
print train_data.count()
print train_data.T.apply(lambda x: x.nunique(),axis=1)
```

```python
train, test = train_test_split(train_data, test_size = 0.2)
test=test.dropna()
train=train.dropna()
train_data = train
train_data.describe()
```

```python
trip_type_count={}

for x in train_data.TripType:
    if x not in trip_type_count:
        trip_type_count[x]=1
    else:
        trip_type_count[x]+=1
print trip_type_count.items()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 5))
plt.bar(range(len(trip_type_count)), trip_type_count.values(), align='center')
plt.xticks(range(len(trip_type_count)), trip_type_count.keys())

plt.show()
```

```python
plt.figure(figsize=(12,5))
#train_data.groupby(['ScanCount']).size().plot('bar',logy='True')
train_data['ScanCount'].value_counts().plot('bar',logy='True',title="Scan Count Histogram") 
```

```python
plt.figure(figsize=(12,5))
train_data['Weekday'].value_counts().plot('bar',title="Daywise visits in a week ")

def transform(day,no):
    global train_data
    train_data=train_data.replace(day,no)
    return;
transform("Monday", 1)
transform("Tuesday", 2)
transform("Wednesday", 3)
transform("Thursday", 4)
transform("Friday", 5)
transform("Saturday", 6)
transform("Sunday", 7)
train_data=train_data.replace("MENS WEAR","MENSWEAR")
```
